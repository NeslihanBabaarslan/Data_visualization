1. Function Method

    *plt.plot(x,y) grafik synax
    
    *plt.xlabel("") x label
    
    *plt.ylabel(

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
```

2. Object oriented method
 * fig, ax=plt.subplots(figsize=())
 * ax.plot(x,y, c="r")
 * ax.set_xlabel("age")
 * ax.set_ylabel("salary")
 * ax.set_title("Salary by Age")


```python
fig,ax=plt.subplots()  #object oriented
ax.plot(age, salary,"r")
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_title("Salary by Age");
```

```python
fig, ax=plt.subplots(1,2)
ax[0].plot(x,y)
ax[1].plot(x,x**2);
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

 Not: Function method ile coklu grafik çiziminde "plt.subplot(2,1,1)"
        kullanılır iken object oriented "fig, ax=plt.subplots(1,2)" kullanılır.

```python
plt.subplot(2,1,1)
plt.plot(age, salary, "r")
plt.xlabel("age vs salary")
plt.ylabel("salary")

plt.subplot(2,1,2)
plt.plot(age, salary_2, "b")
plt.xlabel("age vs salary_2")
plt.ylabel("salary")
plt.tight_layout();
```

```python
plt.subplot(1,2,1)
plt.plot(age, salary, "r")
plt.subplot(1,2,2)
plt.plot(age, salary_2, "b")
plt.tight_layout();
```

```python
plt.plot(age, salary, "r", label="x")
plt.plot(age, salary_2, "b", label="y")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("age vs salary")
plt.legend();
```

```python
x = np.arange(1,16)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4)

ax[1].plot(x,x**2,x,x**3,x,x**4)
ax[1].set_xlim([2,4])   # zoom yapma işi  limit işi
ax[1].set_ylim([5,100]) 
ax[1].set_ylim([5,100])   # zoom yapma işi

ax[0].axvline(x=3,ls="--")  # zoom yaptım sağda da gösterdim
ax[0].axhline(y=50,ls="--")

ax[1].axvline(x=3,ls="--")  # zoom yaptığm yeri sağda da gösterdim kırılmaları gör
ax[1].axhline(y=50,ls="--");
```
